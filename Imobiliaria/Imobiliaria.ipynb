<h1>Colunas Numericas</h1>
<h2>Valor por mes dos imoveis</h2>

```python
dados = dados.fillna(0)
dados['Valor_por_mes'] = dados.Valor + dados.Condominio
dados.head()
```

<h2>Valor por ano dos imoveis</h2>

```python
dados['Valor_por_ano'] = dados.Valor_por_mes * 12 + dados.IPTU
dados.head()
```

<h1>Colunas categoricas</h1>

<h2>Descricao</h2>

```python
dados['Descricao'] = dados.Tipo + ' em ' + dados.Bairro + ', com ' + \
                    dados.Quartos.astype(str) + ' quarto(s) e ' + \
                    dados.Vagas.astype(str) + ' vaga(s)'
```

```python
dados.head()
```

<h2>Coluna Binaria</h2>

```python
dados['possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Nao")
```

```python
dados.head()
```

```python
dados.to_csv('dados_completos.csv', index=False, sep=';')
```
