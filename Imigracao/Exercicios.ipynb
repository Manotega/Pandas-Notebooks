```python
import pandas as pd

df = pd.read_csv('imigrantes_canada.csv')
```

```python
df.set_index("Pa√≠s", inplace=True)
anos = list(map(str, range(1980, 2014)))
brasil = df.loc['Brasil', anos]
argentina = df.loc['Argentina', anos]
```

```python
dicionario = {
    'anos': brasil.index.tolist(),
    'Brasil': brasil.values.tolist(),
    'Argentina': argentina.values.tolist()
}
dicionario
```

```python
df = pd.DataFrame(dicionario)
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title('Imigracao para o Canada')
plt.xlabel('Ano')
plt.ylabel('Numero de imigrantes')
plt.plot(df['anos'], df['Brasil'], df['Argentina'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2013'])
plt.legend(['Brasil', 'Argentina'])
plt.show()
```

<h1>LOJAS</h1>

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

vendas = pd.DataFrame(vendas_2022, index=lojas)
vendas
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(2, 2, figsize=(20,10))
fig.subplots_adjust(hspace=0.3)
fig.suptitle('Vendas das principais lojas em 2022')

axs[0, 0].plot(vendas.loc['A'], color='blue')
axs[0, 0].set_title('Loja A')

axs[0, 1].plot(vendas.loc['B'], color='red')
axs[0, 1].set_title('Loja B')

axs[1, 0].plot(vendas.loc['C'], color='green')
axs[1, 0].set_title('Loja C')

axs[1, 1].plot(vendas.loc['D'], color='yellow')
axs[1, 1].set_title('Loja D')

ymin = 0
ymax = 450
for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

for ax in axs.flat:
    ax.set_xlabel('Mes')
    ax.set_ylabel('Total de vendas')

plt.show()
```
