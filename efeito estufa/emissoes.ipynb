```python
populacao = pd.read_excel('POP2022_Municipios.xls', header=1, skipfooter=34)
populacao
```

```python
populacao.groupby('UF').sum(numeric_only=True)
```

```python
# populacao['POPULAÇÃO'].astype(int)
print("Deu erro :o")
```

```python
populacao['POPULAÇÃO'].str.contains('\(', na=False)
```

```python
populacao[populacao['POPULAÇÃO'].str.contains('\(', na=False)]
```

```python
populacao = populacao.assign(populacao_sem_parenteses = populacao['POPULAÇÃO'].replace('\(\d{1,2}\)', '', regex=True),
                            populacao_regular = lambda x: x.loc[:, 'populacao_sem_parenteses'].replace('\.', '', regex=True))
```

```python
populacao[populacao['POPULAÇÃO'].str.contains('\(', na=False)]
```

```python
populacao = populacao.astype({'populacao_regular':'int64'})
# ou
populacao['populacao_regular'] = populacao['populacao_regular'].astype('int64')
```

```python
populacao['populacao_regular']
```

```python
populacao_estados = populacao.groupby('UF')[['populacao_regular']].sum().reset_index()
populacao_estados
```

```python
emissao_estados = emissoes_por_ano[emissoes_por_ano['Ano'] == 2021].groupby('Estado')[['Emissão']].sum().reset_index()
emissao_estados
```

```python
dados_agrupados = pd.merge(populacao_estados, emissao_estados, right_on = 'Estado', left_on = 'UF')
dados_agrupados
```

```python
dados_agrupados.plot(x = 'populacao_regular', y = 'Emissão', kind = 'scatter', figsize = (8,6));
```

```python
import plotly.express as px
```

```python
px.scatter(data_frame = dados_agrupados, x = 'populacao_regular', y = 'Emissão', text = 'UF', opacity = 0)
```

```python
dados_agrupados = dados_agrupados.assign(emissao_per_capita = dados_agrupados['Emissão'] / dados_agrupados['populacao_regular']).sort_values('emissao_per_capita', ascending = False)
dados_agrupados = dados_agrupados.drop('Estado', axis = 1)
dados_agrupados
```

```python
px.bar(data_frame = dados_agrupados, x = 'UF', y = 'emissao_per_capita')
```

```python
px.scatter(data_frame = dados_agrupados, x = 'populacao_regular', y = 'Emissão', text = 'UF', size = 'emissao_per_capita')
```
