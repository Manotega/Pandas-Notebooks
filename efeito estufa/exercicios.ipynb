```python
colunas_info = list(dados.loc[:, 'Nível 1 - Setor':'Produto'].columns)
anos = list(dados.loc[:, 1970:2021].columns)
dados = dados.melt(id_vars = colunas_info, value_vars=anos, var_name='Ano', value_name='Emissão')
```

```python
dados.groupby(['Estado', 'Nível 1 - Setor'])[['Emissão']].sum()
```

```python
emissao_por_estado = dados.groupby(['Estado', 'Nível 1 - Setor'])[['Emissão']].sum()
```

```python
emissao_por_estado.xs('Energia', level=1)
```

```python
emissao_por_estado.xs('MG', level=0).idxmax()
```

```python
emissao_por_estado.groupby(level = 0).idxmax()
emissao_por_estado.groupby(level=1).idxmax()
```

```python
max_por_estado = dados.groupby(['Estado', 'Nível 1 - Setor'])[['Emissão']].max()
max_por_estado
```

```python
max_por_estado.pivot_table(index='Estado', columns='Nível 1 - Setor', values='Emissão')
```
