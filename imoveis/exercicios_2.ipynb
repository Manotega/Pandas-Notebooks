```python
import pandas as pd
url = 'https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json'
```

```python
imoveis = pd.read_json(url)
imoveis
```

```python
imoveis = pd.json_normalize(imoveis['dados_locacao'])
imoveis
```

```python
colunas = list(imoveis.columns)
colunas
```

```python
imoveis = imoveis.explode(colunas[1:])
imoveis.reset_index(inplace = True, drop = True)
imoveis
```

```python
imoveis['valor_aluguel'] = imoveis['valor_aluguel'].apply(lambda x: x.replace('$', '').replace(',', '.').replace('reais', '').strip()).astype('float64')
imoveis
```

```python
imoveis['apartamento'].str.replace(r'\(blocoAP\)', '', regex=True).str.strip()
```

```python
imoveis['apartamento'] = imoveis['apartamento'].str.replace(r'\(blocoAP\)', '', regex=True).str.strip()
imoveis
```

```python
imoveis['datas_combinadas_pagamento'] = pd.to_datetime(imoveis['datas_combinadas_pagamento'], format='%d/%m/%Y')
imoveis['datas_de_pagamento'] = pd.to_datetime(imoveis['datas_de_pagamento'], format='%d/%m/%Y')
```

```python
imoveis.info()
```

```python
imoveis['atraso'] = (imoveis['datas_de_pagamento'] - imoveis['datas_combinadas_pagamento']).dt.days
imoveis
```

```python
atraso_por_apartamento = imoveis.groupby('apartamento')['atraso'].sum().sort_values(ascending=False)
atraso_por_apartamento
```

```python
media_atraso = imoveis.groupby('apartamento')['atraso'].mean().sort_values(ascending=False)
media_atraso
```
