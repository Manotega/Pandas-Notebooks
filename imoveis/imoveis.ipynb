```python
import pandas as pd
```

```python
url_1 = "https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_hospedagem.json"
url_2 = "https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/moveis_disponiveis.json"
```

```python
dados = pd.read_json(url_1)
dados
```

```python
dados = pd.json_normalize(dados['info_moveis'])
dados
```

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[3:])
dados
```

```python
dados.reset_index(inplace = True, drop = True)
dados.head()
```

```python
dados.info()
```

```python
dados['max_hospedes'] = dados['max_hospedes'].astype('int64')
```

```python
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
```

```python
dados[col_numericas] = dados[col_numericas].astype('int64')
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype('float64')
```

```python
dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip()).astype('float64')
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].map(lambda x: x.replace('$', '').replace(',', '').strip()).astype('float64')
```

```python
dados['comodidades'] = dados['comodidades'].str.replace(r'[\"\{\}]', '', regex = True)
dados.head()
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower()
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace(r'[^a-zA-Z0-9\-\']', ' ', regex = True)
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace(r'(?<!\w)-(?!\w)', ' ', regex = True)
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split()
dados.head()
```
