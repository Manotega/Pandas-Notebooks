```python
import pandas as pd

url_1 = 'https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_vendas_clientes.json'
```

```python
vendas_clientes = pd.read_json(url_1)
vendas_clientes
```

```python
vendas_clientes = pd.json_normalize(vendas_clientes['dados_vendas'])
vendas_clientes
```

```python
colunas_vendas = list(vendas_clientes.columns)
vendas_clientes = vendas_clientes.explode(colunas_vendas[1:])
vendas_clientes
```

```python
vendas_clientes.info()
```

```python
vendas_clientes['Valor da compra'].apply(lambda x: x.replace('R$', '').replace(',', '.').strip()).astype('float64')
```

```python
vendas_clientes.reset_index(inplace = True, drop = True)
vendas_clientes
```

```python
vendas_clientes['Valor da compra'] = vendas_clientes['Valor da compra'].apply(lambda x: x.replace('R$', '').replace(',', '.').strip()).astype('float64')
vendas_clientes
```

```python
vendas_clientes['Cliente'].str.lower()
```

```python
vendas_clientes['Cliente'] = vendas_clientes['Cliente'].str.lower()
vendas_clientes['Cliente'].str.replace(r'[^a-zA-Z]', ' ', regex = True).str.strip()
```

```python
vendas_clientes['Cliente'] = vendas_clientes['Cliente'].str.replace(r'ã', 'a', regex = True).str.strip()
vendas_clientes['Cliente'] = vendas_clientes['Cliente'].str.replace(r'[^a-z]', ' ', regex = True).str.strip()
vendas_clientes
```

```python
vendas_clientes['Data de venda'] = pd.to_datetime(vendas_clientes['Data de venda'], format = '%d/%m/%Y')
vendas_clientes
```

```python
valores_clientes = vendas_clientes.groupby('Cliente')['Valor da compra'].sum().sort_values(ascending = False)
valores_clientes
```
